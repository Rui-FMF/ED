# Univariate Analysis

```python
import pandas

testDataframe = pandas.read_csv('test_FD001.csv', sep=';')
trainDataframe = pandas.read_csv('train_FD001.csv', sep=';')

testDataframeStats = testDataframe.agg(func=['max','min','mean','median','var','std'])
trainDataframeStats = testDataframe.agg(func=['max','min','mean','median','var','std'])

testDataframeStats.iloc[:,:-1].to_excel('testStats.xlsx')
trainDataframeStats.iloc[:,:-1].to_excel('trainStats.xlsx')

print('Test Data\n', testDataframe.iloc[:,-1].describe())
print('Train Data\n', trainDataframe.iloc[:,-1].describe())
```
